```
def statistics_for_lists(lst):
  print("Maximum Value Of List:")
  print(max(lst))
  print("Median of the List:")
  print(statistics.median(lst))
  print("Mean of the List:")
  print(statistics.mean(lst))
  print("75% of the Median is:")
  print(statistics.median_high(lst))
  print("Minimum Value of List:")
  print(min(lst))
```

```
deliveries = pd.read_csv("/content/sample_data/deliveries.csv")
deliveries.head()
```

```
dropToBeList = ['inning','is_super_over','bye_runs','legbye_runs','fielder']
deliveries.drop(dropToBeList, axis=1, inplace=True)
deliveries.replace(['Deccan Chargers','Delhi Daredevils'],['Sunrisers Hyderabad','Delhi Capitals'],inplace=True,regex=True)
deliveries['dismissal_kind'].fillna('Not Out',inplace=True)
deliveries.head()
```

```
ballbyball = deliveries[((deliveries.batting_team=="Royal Challengers Bangalore") & (deliveries.bowling_team=='Delhi Capitals')) | ((deliveries.batting_team=='Delhi Capitals') & (deliveries.bowling_team=="Royal Challengers Bangalore"))]
no_of_matches=list(set([i for i in ballbyball['match_id']]))
no_of_matches.sort()
print(len(no_of_matches))
```

```
#Q3
rabada = deliveries[deliveries.bowler=="K Rabada"]
lst_match_ids = list(set([i for i in rabada['match_id']]))
fst_wkt = []
for i in lst_match_ids:
  match_stats = rabada[rabada.match_id==i]
  ball=0
  no_nob_wide = match_stats[(match_stats.wide_runs==0) & (match_stats.noball_runs==0)]
  flag=0
  for i in no_nob_wide['dismissal_kind']:
    if i=='Not Out':
      ball += 1
    else:
      flag=1
      ball += 1
      break
  if(flag==1):
    fst_wkt.append(ball)
print(len(lst_match_ids))
print(len(fst_wkt))
statistics_for_lists(fst_wkt)
```

Out of 18 matches he played, He got wicket in 16 of his matches. That amounts for 89% of his total matches. He gets his wicket around 12th ball according to the statistics

```
#Q2
match_ids = list(set([i for i in ballbyball['match_id']]))
no_of_players = []
for i in match_ids:
  df = ballbyball[ballbyball.match_id==i]
  lst_batsman = list(set([k for k in df['batsman']]))
  count=0
  for j in lst_batsman:
    df_bat = df[df.batsman==j]
    tot_runs = df_bat.groupby('match_id').batsman_runs.sum()
    tot_balls = df_bat.groupby('match_id').ball.count()
    strike_rate = round((tot_runs.values[0]*100)/tot_balls.values[0],2)
    no_of_4s = df_bat[(df_bat.batsman_runs==4)]
    no_of_4s = no_of_4s.shape[0]
    no_of_6s = df_bat[df_bat.batsman_runs==6]
    no_of_6s = no_of_6s.shape[0]
    no_boundaries=no_of_4s + no_of_6s
    if(no_boundaries > 3 and strike_rate>130):
      if(no_of_4s >= 2):
        count += 1 
  no_of_players.append(count)
statistics_for_lists(no_of_players)
```

```
#Q4
match_ids = list(set([i for i in ballbyball['match_id']]))
no_of_players = []
for i in match_ids:
  df = ballbyball[ballbyball.match_id==i]
  lst_bowler= list(set([k for k in df['bowler']]))
  count=0
  for j in lst_bowler:
    bowler_stats = df[df.bowler==j]
    no_of_balls = bowler_stats[(bowler_stats.wide_runs==0) & (bowler_stats.noball_runs==0)]
    no_of_balls = no_of_balls.shape[0]
    dot_balls = bowler_stats[bowler_stats.total_runs==0]
    dot_balls = dot_balls.shape[0]
    runs_given = bowler_stats['total_runs'].sum()
    overs_bowled = bowler_stats['over'].nunique()
    boundaries_gvn = bowler_stats[(bowler_stats.batsman_runs==4) | (bowler_stats.batsman_runs==6)]
    boundaries_gvn = boundaries_gvn.shape[0]
    economy_rate = round(runs_given/overs_bowled, 2)
    if(overs_bowled > 2):
      if(boundaries_gvn<4 and economy_rate<8 and dot_balls>8):
        count += 1
  no_of_players.append(count)
print(no_of_players)
statistics_for_lists(no_of_players)
```
