```
#Q2
bis_mi = []
bis_rr = []
mi_lineup = list(set([i for i in mi_batting_stats['batsman']]))
rr_lineup = list(set([i for i in rr_batting_stats['batsman']]))
for i in mi_lineup:
  batsman_stats =  mi_batting_stats[mi_batting_stats['batsman']==i]
  total_runs_scored = batsman_stats.groupby('match_id').batsman_runs.sum().reset_index()
  total_balls_faced = batsman_stats.groupby('match_id').ball.count().reset_index()
  nof_boundaries = batsman_stats[batsman_stats['batsman_runs']==4].groupby('match_id').batsman_runs.count().reset_index()
  nof_boundaries.rename(columns={'batsman_runs':'nof_4s'}, inplace=True)
  nof_one_two_three = batsman_stats[(batsman_stats['batsman_runs']==1)|(batsman_stats['batsman_runs']==2)|(batsman_stats['batsman_runs']==3)].groupby('match_id').batsman_runs.count().reset_index()
  nof_one_two_three.rename(columns={'batsman_runs':'nof_1_2_3'},inplace=True)
  df1 = pd.merge(total_runs_scored,total_balls_faced,how='outer')
  df2 = pd.merge(nof_boundaries,nof_one_two_three,how='outer')
  df = pd.merge(df1,df2,how='outer')
  df.fillna(0,inplace=True)
  df = df[df['ball']>10]  
  val1 = sum(df['batsman_runs'])+sum(df['nof_4s'])*4
  val2 = sum(df['nof_1_2_3'])*0.5 + sum(df['ball'])
  bis_mi.append(val1-val2)
for i,j in zip(mi_lineup,bis_mi):
  print(i,end=" ")
  print(j)
  print("\n")
for i in rr_lineup:
  batsman_stats =  rr_batting_stats[rr_batting_stats['batsman']==i]
  total_runs_scored = batsman_stats.groupby('match_id').batsman_runs.sum().reset_index()
  total_balls_faced = batsman_stats.groupby('match_id').ball.count().reset_index()
  nof_boundaries = batsman_stats[batsman_stats['batsman_runs']==4].groupby('match_id').batsman_runs.count().reset_index()
  nof_boundaries.rename(columns={'batsman_runs':'nof_4s'}, inplace=True)
  nof_one_two_three = batsman_stats[(batsman_stats['batsman_runs']==1)|(batsman_stats['batsman_runs']==2)|(batsman_stats['batsman_runs']==3)].groupby('match_id').batsman_runs.count().reset_index()
  nof_one_two_three.rename(columns={'batsman_runs':'nof_1_2_3'},inplace=True)
  df1 = pd.merge(total_runs_scored,total_balls_faced,how='outer')
  df2 = pd.merge(nof_boundaries,nof_one_two_three,how='outer')
  df = pd.merge(df1,df2,how='outer')
  df.fillna(0,inplace=True)
  df = df[df['ball']>10]
  val1 = sum(df['batsman_runs'])+sum(df['nof_4s'])*4
  val2 = sum(df['nof_1_2_3'])*0.5 + sum(df['ball'])
  bis_rr.append(val1-val2)
for i,j in zip(rr_lineup,bis_rr):
  print(i,end=" ")
  print(j)
  print("\n")
```

```
#Q5
bls_mi = []
bls_rr = []
mi_bowling_stats = deliveries[deliveries['bowling_team']=='Mumbai Indians']
rr_bowling_stats = deliveries[deliveries['bowling_team']=='Rajasthan Royals']
mi_lineup_b = list(set([i for i in mi_bowling_stats['bowler']]))
rr_lineup_b = list(set([i for i in rr_bowling_stats['bowler']]))
for i in mi_lineup_b:
  bowler_stats =  mi_bowling_stats[mi_bowling_stats['bowler']==i]
  wickets_taken = bowler_stats[(bowler_stats.dismissal_kind=='caught')|(bowler_stats.dismissal_kind=='bowled')|(bowler_stats.dismissal_kind=='lbw')].groupby('match_id').dismissal_kind.count().reset_index()
  wickets_taken.rename(columns={'dismissal_kind':'no_of_wickets'},inplace=True)
  dot_balls = bowler_stats[bowler_stats['total_runs']==0].groupby('match_id').total_runs.count().reset_index()
  dot_balls.rename(columns={'total_runs':'dot_balls'},inplace=True)
  nof_boundaries = bowler_stats[bowler_stats['batsman_runs']==4].groupby('match_id').batsman_runs.count().reset_index()
  nof_boundaries.rename(columns={'batsman_runs':'nof_4s_given'}, inplace=True)
  total_balls = bowler_stats.groupby('match_id').ball.count().reset_index()
  total_balls.rename(columns={'ball':'no_of_balls'},inplace=True)
  df1 = pd.merge(wickets_taken,dot_balls,how='outer')
  df2 = pd.merge(nof_boundaries,total_balls,how='outer')
  df = pd.merge(df1,df2,how='outer')
  df.fillna(0,inplace=True)  
  df['no_of_balls'].replace([7,8,9,10],6,inplace=True)
  df['no_of_balls'].replace([13,14,15],12,inplace=True)
  df['no_of_balls'].replace([19,20,21],18,inplace=True)
  df['no_of_balls'].replace([25,26,27],24,inplace=True)
  val1 = sum(df['no_of_wickets'])+sum(df['dot_balls'])*2
  val2 = sum(df['nof_4s_given'])/((sum(df['no_of_balls'])-sum(df['dot_balls']))*2.5)
  val=val1-val2
  bls_mi.append(val)
for i,j in zip(mi_lineup_b,bls_mi):
  print(i,end=" ")
  print(j)
  print("\n")
```
