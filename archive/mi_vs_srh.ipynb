```
deliveries = pd.read_csv("/content/sample_data/deliveries.csv")
deliveries.head()
```

```
dropToBeList = ['inning','is_super_over','bye_runs','legbye_runs','fielder']
deliveries.drop(dropToBeList, axis=1, inplace=True)
deliveries.replace(['Deccan Chargers','Delhi Daredevils'],['Sunrisers Hyderabad','Delhi Capitals'],inplace=True,regex=True)
deliveries['dismissal_kind'].fillna('Not Out',inplace=True)
deliveries.head()
```

```
ballbyball = deliveries[((deliveries.batting_team=="Mumbai Indians") & (deliveries.bowling_team=='Sunrisers Hyderabad')) | ((deliveries.batting_team=='Sunrisers Hyderabad') & (deliveries.bowling_team=="Mumbai Indians"))]
no_of_matches=list(set([i for i in ballbyball['match_id']]))
no_of_matches.sort()
print(len(no_of_matches))
```

```
#Q5
no_of_wickets_fall = ballbyball[ballbyball.dismissal_kind != 'Not Out'].groupby('match_id').dismissal_kind.count()
no_of_wickets_fall.describe()
```

There is on an average, chance of fall of 12 to 15 wickets in this match

```
#Q3
lst_batsman = list(set([i for i in ballbyball['batsman']]))
smash_ratios = []
for i in lst_batsman:
  ratios=[]
  df = ballbyball[ballbyball.batsman==i]
  no_of_balls_faced = df[(df.wide_runs==0) | (df.noball_runs==0)].groupby('match_id').ball.count().reset_index()
  boundaries_scored = df[(df.batsman_runs==4) | (df.batsman_runs==6)].groupby('match_id').batsman_runs.count().reset_index()
  d = pd.merge(no_of_balls_faced, boundaries_scored, how='outer')
  d.fillna(0, inplace=True)
  balls_lst = [k for k in d['ball']]
  boundaries_lst = [k for k in d['batsman_runs']]
  for j in range(len(balls_lst)):
    ratios.append((boundaries_lst[j]*100)/balls_lst[j])
  smash_ratios.append(statistics.median(ratios))

for i,j in zip(lst_batsman,smash_ratios):
  print(i+" "+str(j))
```

```
#Q4
lst_bowlers = list(set([i for i in ballbyball['bowler']]))
boundary_leaker=[]
for i in lst_bowlers:
  df = ballbyball[ballbyball.bowler==i]
  balls_bowled = df[(df.wide_runs==0) | (df.noball_runs==0)].groupby('match_id').ball.count().reset_index()
  no_of_boundaries = df[(df.batsman_runs==4) | (df.batsman_runs==6)].groupby('match_id').batsman_runs.count().reset_index()
  no_of_boundaries.rename(columns={'batsman_runs':'no_of_4_6'},inplace=True)
  d = pd.merge(no_of_boundaries,balls_bowled,how='outer')
  d.fillna(0,inplace=True)
  d['no_of_4_6'] += 1
  d['val'] = round(d['no_of_4_6']/d['ball'],2)
  d['reqVal'] = round(1/d['val'],2)
  boundary_leaker.append(statistics.median(d['reqVal']))

for i,j in zip(lst_bowlers,boundary_leaker):
  print(i+" "+str(round(j,2)))
```

```
stamina_for_player=[]
normalized_stamina=[]
for i in lst_batsman:
  df = ballbyball[ballbyball.batsman==i]
  df_1 = ballbyball[ballbyball.non_striker==i]
  singles = df[df.batsman_runs==1].groupby('match_id').batsman_runs.count().reset_index()
  singles.rename(columns={'batsman_runs':'1s'}, inplace=True)
  fours = df[df.batsman_runs==4].groupby('match_id').batsman_runs.count().reset_index()
  fours.rename(columns={'batsman_runs':'4_runs'}, inplace=True)
  fours['4_runs'] *= 4
  total_runs = df.groupby('match_id').batsman_runs.sum().reset_index()
  balls_faced = df[(df.wide_runs==0) | (df.noball_runs==0)].groupby('match_id').ball.count().reset_index()
  balls_non_striker_end = df_1.groupby('match_id').non_striker.count().reset_index()
  balls_non_striker_end.rename(columns={'non_striker':'no_balls_on_crease'}, inplace=True)
  df1 = pd.merge(total_runs, balls_faced, how='outer')
  df2 = pd.merge(df1, singles, how='outer')
  df4 = pd.merge(df2, fours ,how='outer')
  df3 = pd.merge(df4, balls_non_striker_end, how='outer')
  df3 = df3[df3['batsman_runs']>15]
  df3['no_balls_on_crease'] += df3['ball']
  df3['stamina_score'] = round(((df3['1s']/df3['4_runs'])+(df3['ball']/df3['no_balls_on_crease']))/df3['batsman_runs'],2)
  try:
    stamina_for_player.append(statistics.mean(df3['stamina_score']))
  except:
    stamina_for_player.append(0)
req_indexLst = [round(i,3) for i in range(len(stamina_for_player)) if stamina_for_player[i]>0]
req_stamina_val = [stamina_for_player[i] for i in req_indexLst]
req_lst_batsman = [lst_batsman[i] for i in req_indexLst]
for i in range(len(req_stamina_val)):
  val = round((req_stamina_val[i]-min(req_stamina_val))/(max(req_stamina_val)-min(req_stamina_val)),2)
  normalized_stamina.append(val)
```

```
for i,j in zip(req_lst_batsman, normalized_stamina):
  print(i+" "+str(j))
```
